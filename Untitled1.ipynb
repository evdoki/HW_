```python
def basic_calculator(num1, num2):
    try:
        operator = str(input('Введите оператор: '))
        list_oper = ['+','-','*','/']
        return {
        operator == list_oper[0]: int(num1)+int(num2),
        operator == list_oper[1]: int(num1)-int(num2),
        operator == list_oper[2]: int(num1)*int(num2),
        operator == list_oper[3]: int(num1)/int(num2),
        operator not in list_oper: 'Введите корректный оператор'}[1]
    
    except ValueError:
        print('Ошибка преобразования типов')
    except ZeroDivisionError:
        print('Ошибка деления на ноль')
```

```python
basic_calculator(2,3)
```

```python
import string
with open('moby.txt','r') as i_f, open('m2.txt','w') as o_f:
    txt = i_f.readlines()
    char = list(map(lambda x: x.upper(), txt))
    translator = str.maketrans('','',string.punctuation)
    wp = ''.join(list(map(lambda x: x.translate(translator), char)))
    text = '\n'.join(wp.split("\\"))
    o_f.writelines(text)
```

```python
def coun(stroka):
    count = {}
    for word in stroka:
        count.setdefault(word, 0)
        count[word] += 1
        return count
```

```python
with open('m2.txt','r') as f:
    text = f.read()
    clear = text.replace("\\n","\\")
    res = coun(clear.split())
    sortW = sorted(res.items(), key=lambda x: x[1], reverse=True)
    lenL = len(sortW)
    mmax = sortW[0:5]
    mmin = sortW[lenL-5:lenL]
```

```python
def wCSV(name, lst):
    ad = ['name,','address,','age']
    with open(f'{name}.csv','w', encoding="utf8") as of:
        for el in lst: 
            ad.append('\\n'+','.join(el))
            of.write(''.join(ad))
        return f'Файл был сохранён под именем: {name}'
```

```python
def rCSV(name):
    with open(name,'r', encoding="utf8") as f:
        x = f.readlines()
        file = ''.join(x).replace('\\n',',').split(',')
        keys = file[0:3]
        values = []
        for i in range(3,len(file), 3):
            values.append(file[i:i+3])
            return list(map(lambda x: dict(zip(keys, x)), values)) 
```

```python
with open("text.txt",'r', encoding='utf8') as f:
    text = f.read()
    print('Cреднее количество слов:', len(text.split()) / len(text.split('\\n')))
    print('Количество букв в файле в верх:', sum(1 for i in text if i.isupper()))
    print('Количество букв в ниж:', sum(1 for i in text if i.islower()))
    print('Количество цифр:', sum(1 for i in text if i.isdigit()))
    print('Количество пробелов:', sum(1 for i in text if i == " "))
```
